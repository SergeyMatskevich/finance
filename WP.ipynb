```python
list_agent_vitalur = ['VITALUR']
list_category_restaurant = ['LIVAN HAUS','PAPA DONER','PT RESTORAN PHO BO','DOLCHELIUKS BRAND STORE',
                            'RBO "SALATEIRA"','RESTORAN "SALATEIRA"','PEKARNYA BULKY','RESTORAN KAPARA BPSB',
                            'RBO "BURGER KING"','Chayhana Baklazhan']
list_category_food = ['SHOP "EVROOPT" MVV','SUPERMARKET "GREEN" BAPB','SHOP "VITALUR"','KUSOCHEK SCHASTYA','I.-SHOP EVROOPT MVV']
list_category_electronics = ['IMARKET.BY','I.-SHOP WWW.SILA.BY']
list_category_pharmacy = ['FITOBEL APTEKA 29']
list_category_transport = ['TRANSPORT PAYBYCARD.BY BP','WWW.KOLOBIKE.BY']
list_category_grocery = ['PT OAO TSUM MINSK','VDOM SHOP MINSK','YORK KHOZTOVARY']
list_category_fx = ['В/о оп ерация по курсу Банка через  АК','покуп ка долларов США по курсу']
list_category_insync = ['InSync (ERIP)']
list_category_clothes = ['WWW.WILDBERRIES.B','M-n Conte']
list_category_toys = ['PT SUPERMARKET DLYA DETEJ']
list_category_atm = ['ATMALF','ATMBSB']
list_category_salary = ['Зачисление Заработная плата']
```

```python
def trx_category(row):
    if any(ext in row['Operation'] for ext in list_category_restaurant):
        return 'restaurant'
    if any(ext in row['Operation'] for ext in list_category_food):
        return 'food'
    if any(ext in row['Operation'] for ext in list_category_electronics):
        return 'electronics'
    if any(ext in row['Operation'] for ext in list_category_pharmacy):
        return 'pharmacy'
    if any(ext in row['Operation'] for ext in list_category_transport):
        return 'transport'
    if any(ext in row['Operation'] for ext in list_category_grocery):
        return 'grocery'
    if any(ext in row['Operation'] for ext in list_category_fx):
        return 'fx'
    if any(ext in row['Operation'] for ext in list_category_insync):
        return 'insync'
    if any(ext in row['Operation'] for ext in list_category_clothes):
        return 'clothes'
    if any(ext in row['Operation'] for ext in list_category_toys):
        return 'toys'
    if any(ext in row['Operation'] for ext in list_category_atm):
        return 'atm'
    if any(ext in row['Operation'] for ext in list_category_salary):
        return 'salary'
    return 'empty'
```

```python
def format(df):
    df = df[df['Currency'].notna()]
    df = df[df['Currency'].isin(['USD','BYN','RUB'])]
    df = df[df['Fee/Money-back'] != 'USD']
    df = df[df['Fee/Money-back'] != 'BYN']
    df = df[df['Fee/Money-back'] != 'RUB']
    #df['TransactionDate'] = df['TransactionDate'][0]
    df['TransactionDateFormat'] = pd.to_datetime(df['TransactionDate'], format='%d.%m.%Y %H:%M:%S')
    df['AccountTransactionDateFormat'] = pd.to_datetime(df['AccountTransactionDate'], format='%d.%m.%Y')
    #df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
    #df['AccountTransactionDate'] = pd.to_datetime(df['AccountTransactionDate'])
    df['Amount'] =df['Amount'].str.replace(' ', '')
    df['Amount'] = df['Amount'].str.replace(',', '.')
    df['Amount'] = pd.to_numeric(df['Amount'])
    df['AccountAmmount'] = df['AccountAmmount'].str.replace(' ', '')
    df['AccountAmmount'] = df['AccountAmmount'].str.replace(',', '.')
    df['AccountAmmount'] = pd.to_numeric(df['AccountAmmount'])
    df['Fee/Money-back'] = df['Fee/Money-back'].str.replace(' ', '')
    df['Fee/Money-back'] = df['Fee/Money-back'].str.replace(',', '.')
    df['Fee/Money-back'] = pd.to_numeric(df['Fee/Money-back'])
    return df
```
