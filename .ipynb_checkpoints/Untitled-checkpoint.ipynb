```python
import re
import csv
import pandas as pd
import numpy as np
```

```python
columns = ['TransactionDate','Operation','Amount','Currency','AccountTransactionDate','Fee/Money-back','AccountAmmount','Card','TransactionCategory']
```

Load files:

```python
byn1qS = pd.read_csv('Data/BYN1QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
byn2qS = pd.read_csv('Data/BYN2QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
byn3qS = pd.read_csv('Data/BYN3QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
byn1qT = pd.read_csv('Data/BYN1QT.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
byn2qT = pd.read_csv('Data/BYN2QT.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
byn3qT = pd.read_csv('Data/BYN3QT.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
usd1qS = pd.read_csv('Data/USD1QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
usd2qS = pd.read_csv('Data/USD2QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
usd3qS = pd.read_csv('Data/USD3QS.csv', sep=';',header = None, names=columns, encoding='utf_8',skip_blank_lines=True,error_bad_lines=True)
```

```python
def format(df):
    df = df[df['Currency'].notna()]
    df = df[df['Currency'].isin(['USD','BYN','RUB'])]
    df = df[df['Fee/Money-back'] != 'USD']
    df = df[df['Fee/Money-back'] != 'BYN']
    df = df[df['Fee/Money-back'] != 'RUB']
    #df['TransactionDate'] = df['TransactionDate'][0]
    df['TransactionDateFormat'] = pd.to_datetime(df['TransactionDate'], format='%d.%m.%Y %H:%M:%S')
    df['AccountTransactionDateFormat'] = pd.to_datetime(df['AccountTransactionDate'], format='%d.%m.%Y')
    #df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
    #df['AccountTransactionDate'] = pd.to_datetime(df['AccountTransactionDate'])
    df['Amount'] =df['Amount'].str.replace(' ', '')
    df['Amount'] = df['Amount'].str.replace(',', '.')
    df['Amount'] = pd.to_numeric(df['Amount'])
    df['AccountAmmount'] = df['AccountAmmount'].str.replace(' ', '')
    df['AccountAmmount'] = df['AccountAmmount'].str.replace(',', '.')
    df['AccountAmmount'] = pd.to_numeric(df['AccountAmmount'])
    return df
```

```python
byn1qS = format(byn1qS)
byn2qS = format(byn2qS)
byn3qS = format(byn3qS)
usd1qS = format(usd1qS)
usd2qS = format(usd2qS)
usd3qS = format(usd3qS)
byn1qT = format(byn1qT)
byn2qT = format(byn2qT)
byn3qT = format(byn3qT)
```

```python
byn1qS['Owner']='Sergey'
byn2qS['Owner']='Sergey'
byn3qS['Owner']='Sergey'
usd1qS['Owner']='Sergey'
usd2qS['Owner']='Sergey'
usd3qS['Owner']='Sergey'
byn1qT['Owner']='Tanya'
byn2qT['Owner']='Tanya'
byn3qT['Owner']='Tanya'
```

```python
byn1qS['CardCurrency']='BYN'
byn2qS['CardCurrency']='BYN'
byn3qS['CardCurrency']='BYN'
usd1qS['CardCurrency']='USD'
usd2qS['CardCurrency']='USD'
usd3qS['CardCurrency']='USD'
byn1qT['CardCurrency']='BYN'
byn2qT['CardCurrency']='BYN'
byn3qT['CardCurrency']='BYN'
```

```python
final = pd.concat([byn1qS,byn2qS, byn3qS, usd1qS, usd2qS, usd3qS,byn1qT,byn2qT,byn3qT])
```
